```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib import style, rc
import time
```

```python
style.available
```

```python
style.use("classic")
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)

def animate(i):
    data = pd.read_csv("data/sample_text.txt")
    xar = np.array(data.iloc[:,0])
    yar = np.array(data.iloc[:,1])
    
    ax1.plot(xar, yar)
    
ani = animation.FuncAnimation(fig, animate, frames=100, interval=100)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)

def animate(i):
    data = open('data/sentiment.txt', 'r').read()
    lines = data.split('\n')
    xar = []
    yar = []
    
    x = 0
    y = 0
    for l in lines:
        x += 1
        if "pos" in l:
            y += 1
        elif "neg" in l:
            y -= 1
        xar.append(x)
        yar.append(y)
    
    ax1.clear()
    ax1.plot(xar, yar)
    
ani = animation.FuncAnimation(fig, animate, frames=100, interval=100)
ani
```
