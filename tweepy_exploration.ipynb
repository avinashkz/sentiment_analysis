# Basic Twitter

```python
class listener(StreamListener):
    
    def on_data(self,data):
        try:
            #Read the data in as JSON
            data = json.loads(data)
            #Extract only the tweet
            tweet = data['text']
            #Print the tweet
            print(tweet)
            #Save the Tweet with time stamp
            saveTweet = str(time.time()) + '::' + tweet
            
            #Save the tweets in a file
            saveFile = open("data/twitter_data.csv", "a")
            saveFile.write(saveTweet)
            saveFile.write('\n')
            saveFile.close()
            return True
        except BaseException as e:
            print("failed ondata,", str(e))
            time.sleep(5)
    
    def on_error(self, status):
        print(status)

auth = OAuthHandler(ckey, csecret)
auth.set_access_token(atoken, asecret)
twitterStream = Stream(auth, listener())
twitterStream.filter(track = ["obama"])
```

```python
class listener(StreamListener):
    
    def on_data(self,data):
        try:
            #Read the data in as JSON
            data = json.loads(data)
            #Extract only the tweet
            tweet = data['text']
            #Print the tweet
            
            sentiment, confidence = sa.sentiment_module(tweet)
            print(tweet, sentiment, confidence)
            
            if confidence >= 0.6:
                saveFile = open("data/sentiment.txt", "a")
                saveFile.write(sentiment[0])
                saveFile.write('\n')
                saveFile.close()
                
            return True
        
        except BaseException as e:
            print("failed ondata,", str(e))
            time.sleep(5)
    
    def on_error(self, status):
        print(status)

auth = OAuthHandler(ckey, csecret)
auth.set_access_token(atoken, asecret)
twitterStream = Stream(auth, listener())
twitterStream.filter(track = ["happy"])
```
